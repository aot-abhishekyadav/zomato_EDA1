yticklabels = False
Hides row names on the Y-axis
Useful when dataset is large (cleaner look)

cbar = False
Disables the color bar on the side
Color bar is unnecessary because we only care about missing vs not missing

cmap = 'viridis'
Sets the color theme
'viridis' is a smooth color gradient:
Dark color â No missing value
Light color â Missing value

```python
# by heatmap finding null value
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis') # there are many small value of nan value therefore we can't see anything
```

```python
df2 = pd.read_excel('/content/drive/MyDrive/Country-Code[1].xlsx')
df2.head()
```

```python
# we can merge df and df2 dataset
df3 = pd.merge(df,df2, on = 'Country Code', how = 'left')
df3
```

```python
df3.dtypes
```

```python
df3['Country'].value_counts()
```

```python
country_name = df3['Country'].value_counts().index # by the help of index sai hume pta chl jaata hai konsi konsi country hai country column mai aur usse hum name mai store kar rhe hai
country_name
```

```python
country_val = df3['Country'].value_counts().values # values ki help sai hum konsi country ki kitni value hai yai pta chl jaata hai aur usse hum value mai store kar rhe hai
country_val
```

```python
# making piechart of top 3 countries that use zomato with the help of name and value
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%',shadow = True) # [:3] means starting top 3 value of name and value
plt.show()
```

```python
# Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms
```

```python
df3.columns
```

```python
df3.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = df3.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0 : 'Rating Count'})
ratings
# by groupby,size and reset_index sai hum inn 3 column kaa alg sai dataset bana rhe hai main reset_index humare default index jo bhi aata hai use shi sai arrange ka deta hai
```

```python
'''
when rating is between 4.5 to 4.9 = Excellent
when  rating is between 4.0 to 4.4 = very good
when rating is between 3.5 to 3.9 = good
when rating is betwwen 2.5 to 3.5 = Average
when rating is between 1.8 to 2.4 = poor
'''
```

```python
ratings.head(2)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = [12,6] # by this we can increase the size of figure
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings, hue = 'Rating color')
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings, hue = 'Rating color', palette=['blue','red','orange','yellow','green','brown'])
```

```python
sns.countplot(x = 'Rating color', data = ratings,hue = 'Rating color', palette=['blue','red','orange','yellow','green','brown'])
```

```python
# Find the countries name that has given 0 ratings
df3[df3['Rating color'] == 'White'].groupby('Country').size().reset_index(name = 'zero ratings')
```

```python
# 2nd Method
df3.groupby(['Aggregate rating', 'Country']).size().reset_index(name = 'zero ratings').head(4)
```

```python
# which currency is used by which country
df3.groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Which Countries do have online deliveries option
df3[df3['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index(name = 'Online Delivery')
```

```python
df3.groupby(['Country', 'Has Online delivery']).size().reset_index(name = 'Online Delivery')
```
